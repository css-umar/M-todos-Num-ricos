```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

fx=lambda x:25*x**2 - 10*x + 1
x=np.linspace(0,1)

x0=1.0
x1=0.8
fx0=fx(x0)
fx1=fx(x1)

x2=x0-((x1-x0)*fx(x0)/(fx(x1)-fx(x0)))
x3=x1-((x2-x1)*fx(x1)/(fx(x2)-fx(x1)))
x4=x2-((x3-x2)*fx(x2)/(fx(x3)-fx(x2)))
```

```python
plt.figure(figsize=(12,8))
plt.axhline(linewidth=1, color='k')
plt.axvline(linewidth=1, color='k')

plt.plot(x,fx(x))

plt.plot([x0,x1], [fx0,fx1], 'o', markersize=10)
plt.plot([x2], [fx(x2)], 'o', markersize=10)
plt.plot([x3], [fx(x3)], 'o', markersize=10)


plt.plot([x0,x1,x2], [fx0,fx1,0], '--', markersize=10)
plt.plot([x1,x2,x3], [fx1,fx(x2),0], '--', markersize=10)
plt.plot([x2,x3,x4], [fx(x2),fx(x3),0], '--', markersize=10)

plt.plot([x0,x1,x2,x3],[0,0,0,0],"o",markersize=10 )


plt.text(x0,0, r'$x_0$',fontsize=15)
plt.text(x1,0, r'$x_1$',fontsize=15)
plt.text(x2,0, r'$x_2$',fontsize=15)
plt.text(x3,0, r'$x_3$',fontsize=15)
plt.text(x4,0, r'$x_4$',fontsize=15)

plt.annotate(r"$f(x)=25x^2-10x+1$", xy=(1.0,fx(1.0)), xytext=(0.6, 16),
             arrowprops=dict(facecolor='black', shrink=0.05),fontsize=15)


plt.text(x0,fx0, r'$A=(x_0,f(x_0))$',fontsize=15)
plt.text(x1,fx1, r'$B=(x_1,f(x_1))$',fontsize=15)
plt.text(x2,fx(x2), r'$C=(x_2,f(x_2))$',fontsize=15)
plt.text(x3,fx(x3), r'$D=(x_2,f(x_2))$',fontsize=15)

plt.xlabel(r"$x$", fontsize=15)
plt.ylabel(r"$f(x)$",fontsize=15)

plt.show()
```
