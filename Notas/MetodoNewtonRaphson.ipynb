# Newton-Raphson

Sea la siguiente función

$$f(x) = x^3+2x^2+10x-20$$

cuyo gráfico es

```
import numpy as np
import matplotlib.pyplot as plt
def f(x):
  return x**3+2*x**2+10*x-20
x = np.linspace(-5,10)
plt.figure(figsize=(12,8))
plt.plot(x,f(x))
plt.axhline(linewidth=1, color='k')
plt.axvline(linewidth=1, color='k')
plt.xlabel(r"$x$", fontsize=15)
plt.ylabel(r"$f(x)$",fontsize=15)
plt.show()
```

**Paso 1**: Vamos a considerar un valor inicial $x_0$. El cual se va a situar en el eje horizontal. Para el caso que nos ocupa, $x_0 = 8.0$

**Paso 2**: Ubicar el punto $(x_0,f(x_0))$ en el grafico.

**Paso 3**: A partir del punto anterior trace una tangente hasta intersectar con el eje $x$.

**Paso 4**: Ubicar el punto de intersección como $x_1$, el cual es una nueva aproximación a $\bar{x}$.

**Paso 5**: Repetir el proceso, pero ahora comenzando con $x_1$ como nuevo valor inicial.

> _El proceso se repite hasta que se satisfaga:_

$$|f(x_i)|\leq\epsilon_1$$
> _o bien_
$$|x_{i+1}-x_i|\leq\epsilon$$

> **Ojo**: Si lo anterior no se cumple en un número máximo de iteraciones (MAXIT), debe reiniciarse el método con un nuevo valor de $x_0$


```
import numpy as np
import matplotlib.pyplot as plt
def fx(x):
  return x**3+2*x**2+10*x-20
def dfx(x):
    return 3*x**2+4*x+10
x = np.linspace(-5,10)
x0=8.0
fx0=fx(x0)
x1=x0-(fx(x0)/dfx(x0))
fx1=fx(x1)
x2=x1-(fx(x1)/dfx(x1))
fx2=fx(x2)
plt.figure(figsize=(12,8))
plt.axhline(linewidth=1, color='k')
plt.axvline(linewidth=1, color='k')
plt.plot(x,fx(x))
plt.plot([x0,x1,x2], [fx0,fx1,fx2], 'o')
plt.plot([x0,x1], [fx0,0], '--k')
plt.plot([x1,x2], [fx1,0], '--k')
plt.plot([x0,x1,x2,1.36881], [0,0,0,0], 'o')


plt.plot([x0,x0], [0,fx0], '-k')
plt.plot([x1,x1], [0,fx1], '-k')


plt.annotate(r'$f(x)$', xy=(10,fx(10)), xytext=(6, 1000),
             arrowprops=dict(facecolor='black', shrink=0.05),fontsize=15)
plt.annotate(r"$Pendiente=f^{'}(x)=\dfrac{f(x)}{\Delta x}$", xy=(7,420), xytext=(2, 800),
             arrowprops=dict(facecolor='black', shrink=0.05),fontsize=15)
plt.annotate("Raiz real", xy=(1.36881,0), xytext=(-2, 200),
             arrowprops=dict(facecolor='black', shrink=0.05),fontsize=15)


plt.text(1.36881,0, r'$\bar{x}$',fontsize=15)
plt.text(x0,0, r'$x_0$',fontsize=15)
plt.text(x0,fx0, r'$(x_0,f(x_0))$',fontsize=15)
plt.text(x1,0, r'$x_1$',fontsize=15)
plt.text(3,fx1, r'$(x_1,f(x_1))$',fontsize=15)
plt.text(x2,0, r'$x_2$',fontsize=15)
plt.xlabel(r"$x$", fontsize=15)
plt.ylabel(r"$f(x)$",fontsize=15)
plt.show()
```

De la primer iteración se puede obtener la ecuación del algoritmo:

> 1. La pendiente de la tangente en el punto $(x_0,f(x_0))$ es

$$f'(x_0) = \dfrac{f(x_0)}{\Delta x} $$

> 2. Resolviendo para $\Delta x$

$$\Delta x = \dfrac{f(x_0)}{f'(x_0)}$$

> 3. La diferencia entre la nueva aproximación $x_1$ y el valor inicial $x_0$  es $\Delta x$, por lo tanto

$$x_1 = x_0- \dfrac{f(x_0)}{f'(x_0)}$$

## Fallas del método de Newton-Raphson

### No hay raíces reales

Sea la función 

$$f(x) = x^2+2$$

cuya gráfica es la siguiente:

```
get_ipython().magic('reset -sf')    # Borramos memoria
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return x**2+2

def df(x):
  return 2*x

x = np.linspace(-3,3)
plt.figure(figsize=(12,8))
plt.axhline(linewidth=1, color='k')
plt.axvline(linewidth=1, color='k')
plt.plot(x,f(x))
plt.text(-1,4, "Como se puede observar\nLa función no tiene \nraíces reales",fontsize=15)
plt.show()
```

Se observa en la gráfica que la curva de la función nunca se cruza con el eje $x$, por lo tanto las raíces de esta función en particular son complejas.

Si se aplica el método de Newton-Raphson a esta función, el método oscilaría por debajo de la parábola.

```
x0 = -1.0
x1 = x0 - f(x0)/df(x0)
x2 = x1 - f(x1)/df(x1)
x3 = x2 - f(x2)/df(x2)
x4 = x3 - f(x3)/df(x3)

plt.figure(figsize=(12,8))
plt.axhline(linewidth=1, color='k')
plt.axvline(linewidth=1, color='k')

plt.plot(x,f(x))
plt.plot([x0,x1,x2,x3,x4], [0,0,0,0,0], 'o', markersize=12)
plt.plot([x0,x1,x2,x3,x4], [f(x0),f(x1),f(x2),f(x3),f(x4)], 'o',markersize=12)

plt.plot([x0,x0], [0,f(x0)], '--k', markersize=12)
plt.plot([x0,x1], [f(x0),0], '-k', markersize=12)

plt.plot([x1,x1], [0,f(x1)], '--r', markersize=12)
plt.plot([x1,x2], [f(x1),0], '-r', markersize=12)

plt.plot([x2,x2], [0,f(x2)], '--b', markersize=12)
plt.plot([x2,x3], [f(x2),0], '-b', markersize=12)

plt.plot([x3,x3], [0,f(x3)], '--m', markersize=12)
plt.plot([x3,x4], [f(x3),0], '-m', markersize=12)

plt.text(x0,0, r'$x_0$',fontsize=15)
plt.text(x1,0, r'$x_1$',fontsize=15)
plt.text(x2,0, r'$x_2$',fontsize=15)
plt.text(x3,0, r'$x_3$',fontsize=15)
plt.text(x4,0, r'$x_4$',fontsize=15)

plt.text(x0,f(x0), r'$f(x_0)$',fontsize=15)
plt.text(x1,f(x1), r'$f(x_1$)',fontsize=15)
plt.text(x2,f(x2), r'$f(x_2)$',fontsize=15)
plt.text(x3,f(x3), r'$f(x_3)$',fontsize=15)
plt.text(x4,f(x4), r'$f(x_4)$',fontsize=15)

plt.xlabel(r"$x$", fontsize=15)
plt.ylabel(r"$f(x)$",fontsize=15)
plt.show()
```

### Valor inicial muy lejos de la raíz

Sea la función 

$$f(x) = -x^3 + 2x^2 - 2$$

cuya gráfica es la siguiente:

```
get_ipython().magic('reset -sf')    # Borramos memoria
import numpy as np
import matplotlib.pyplot as plt

def f(x):
  return -x**3+2*x**2-2

def df(x):
  return -3*x**2+4*x

x = np.linspace(-2,3)
plt.figure(figsize=(12,8))
plt.axhline(linewidth=1, color='k')
plt.axvline(linewidth=1, color='k')
plt.plot(x,f(x))
plt.annotate("¡Aquí está la raíz!", xy=(-0.9,0), xytext=(-1, 5),
             arrowprops=dict(facecolor='black', shrink=0.05),fontsize=15)

plt.annotate("¡Esto puede ser una trampa!", xy=(1.5,-1), xytext=(1, 10),
             arrowprops=dict(facecolor='black', shrink=0.05),fontsize=15)
plt.show()
```

Obsérvese que la raíz está aproximadamente en $x=-0.9$. 

Para este tipo de situaciones, si se aplica el método de Newton-Raphson con un valor inicial muy alejado de la raíz buscada, por ejemplo $x_0=1.0$, es probable que el método quede atrapado en alguna parte de la función, como se observa en la siguiente gráfica:

```
x0 = 1.0
x1 = x0 - f(x0)/df(x0)
x2 = x1 - f(x1)/df(x1)
x3 = x2 - f(x2)/df(x2)
x4 = x3 - f(x3)/df(x3)

plt.figure(figsize=(12,8))
plt.axhline(linewidth=1, color='k')
plt.axvline(linewidth=1, color='k')

plt.plot(x,f(x))

plt.plot([x0,x1,x2,x3,x4], [0,0,0,0,0], 'o', markersize=12)
plt.plot([x0,x1,x2,x3,x4], [f(x0),f(x1),f(x2),f(x3),f(x4)], 'o',markersize=12)

plt.plot([x0,x0], [0,f(x0)], '--k', markersize=12)
plt.plot([x0,x1], [f(x0),0], '-k', markersize=12)

plt.plot([x1,x1], [0,f(x1)], '--r', markersize=12)
plt.plot([x1,x2], [f(x1),0], '-r', markersize=12)

plt.plot([x2,x2], [0,f(x2)], '--b', markersize=12)
plt.plot([x2,x3], [f(x2),0], '-b', markersize=12)

plt.plot([x3,x3], [0,f(x3)], '--m', markersize=12)
plt.plot([x3,x4], [f(x3),0], '-m', markersize=12)

plt.text(x0,0, r'$x_0$',fontsize=15)
plt.text(x1,0, r'$x_1$',fontsize=15)
plt.text(x2,0, r'$x_2$',fontsize=15)
plt.text(x3,0, r'$x_3$',fontsize=15)
#plt.text(x4,0, r'$x_4$',fontsize=15)

plt.text(x0,f(x0), r'$f(x_0)$',fontsize=15)
plt.text(x1,f(x1), r'$f(x_1$)',fontsize=15)
plt.text(x2,f(x2), r'$f(x_2)$',fontsize=15)
plt.text(x3,f(x3), r'$f(x_3)$',fontsize=15)
#plt.text(x4,f(x4), r'$f(x_4)$',fontsize=15)

plt.xlabel(r"$x$", fontsize=15)
plt.ylabel(r"$f(x)$",fontsize=15)
plt.show()
```

Por el contrario, si se inicia la búsqueda cerca de la raíz, el método converge rápido, como se observa en la siguiente gráfica:

```
x = np.linspace(-1.25,-0.8)
x0 = -1.0
x1 = x0 - f(x0)/df(x0)
x2 = x1 - f(x1)/df(x1)
x3 = x2 - f(x2)/df(x2)
x4 = x3 - f(x3)/df(x3)

plt.figure(figsize=(12,8))
plt.axhline(linewidth=1, color='k')
plt.axvline(linewidth=1, color='k')

plt.plot(x,f(x))
plt.xlim(-1.25,-0.5)

plt.plot([x0,x1,x2], [0,0,0,], 'o', markersize=12)
plt.plot([x0,x1,x2], [f(x0),f(x1),f(x2)], 'o',markersize=12)

plt.plot([x0,x0], [0,f(x0)], '--k', markersize=12)
plt.plot([x0,x1], [f(x0),0], '-k', markersize=12)

plt.plot([x1,x1], [0,f(x1)], '--r', markersize=12)
plt.plot([x1,x2], [f(x1),0], '-r', markersize=12)

plt.plot([x2,x2], [0,f(x2)], '--b', markersize=12)
plt.plot([x2,x3], [f(x2),0], '-b', markersize=12)

plt.text(x0,0, r'$x_0$',fontsize=15)
plt.text(x1,0, r'$x_1$',fontsize=15)
plt.text(x2,0, r'$x_2$',fontsize=15)

plt.text(x0,f(x0), r'$f(x_0)$',fontsize=15)
plt.text(x1,f(x1), r'$f(x_1$)',fontsize=15)
plt.text(x2,f(x2), r'$f(x_2)$',fontsize=15)

plt.xlabel(r"$x$", fontsize=15)
plt.ylabel(r"$f(x)$",fontsize=15)
plt.show()
```

## Método de Newton-Raphson para raíces complejas [1]

Se tiene el siguiente polinomio

$$f(x) = a_n x^n + a_{n-1} x^{n-1} + \ldots + a_1 x + a_0$$

con todos los coeficientes $a_i$ reales. Su derivada $f'(x)$ también es un polinomio con coeficientes reales.

Al aplicar el método de Newton-Raphson a $f(x)$

$$x_1 = x_0 - \dfrac{f(x_0)}{f'(x_0)}$$

+ Si el valor inicial $x_0$ es real, entonces $x_1$ será real. No se puede encontrar una raíz compleja de $f(x)$ si se inicia con $x_0$ real.

+ Si el valor inicial $x_0$ es complejo, $x_1$ sera complejo.

**Ejemplo**

Encuentre las raíces complejas de 

$$f(x) = x^2+4$$

Sea $x_0 = j$, entonces:


```
x0 = 1j    # Valor inicial
EPSILON = 0.0001    # Criterio de paro para la distancia
EPSILON1 = 0.0001    # Criterio de para la función
MAXIT = 40    # Criterio por sino se cumplen las dos anteriores

# Funciones

def f(x):
  return x**2 + 4

def df(x):
  return 2*x

for i in range(MAXIT):
  x1 = x0 - f(x0)/df(x0)   
  d = abs(x1-x0)
  print(i)  
  if d <= EPSILON or abs(f(x1))<=EPSILON1:
    print("--------------")
    print("Iteracion = ", i, "\nSolucion = ",x1)
    break
  else:
    x0 = x1
```

En caso de que una ecuación polinomial tenga raíces complejas con coeficientes reales, éstas aparecen en parejas (complejas conjugadas); es decir,  si $x = a + b j$ es raíz, también lo será $x = a – b j$.

Por lo tanto

$$-2.0j$$

es la segunda raíz de $f(x)=x^2+4$
