```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import optimize
```

```python
def fx(x):
    return x-0.5*(x-3)

def gx(x):
    return 0.5*(x-3)
```

```python
x0=5.0
x1=gx(x0)
x2=gx(x1)
x3=gx(x2)
x4=gx(x3)
x5=gx(x4)
xbar=optimize.newton(fx, 1.5)
```

```python
x=np.linspace(-10,10)
```

```python
plt.figure(figsize=(12,8))
plt.axhline(linewidth=1, color='k')
plt.axvline(linewidth=1, color='k')

plt.plot(x,fx(x), label=r"$f(x)$")
plt.plot(x,gx(x), label=r"$g(x)$")
plt.plot(x,x, label=r"$y=x$")

plt.plot([x0,x1,x2,x3], [0,0,0,0], 'o',markersize=10)
plt.plot([x0,x1,x2,x3], [fx(x0),fx(x1),fx(x2),fx(x3)], 'o',markersize=10)

plt.plot([x0,x0],[0,fx(x0)],"--k", markersize=10)
plt.plot([x0,x1],[fx(x0),0],"--k", markersize=10)
plt.plot([x1,x1],[0,fx(x1)],"--k", markersize=10)
plt.plot([x1,x2],[fx(x1),0],"--k", markersize=10)
plt.plot([x2,x2],[0,fx(x2)],"--k", markersize=10)
plt.plot([x2,x3],[fx(x2),0],"--k", markersize=10)

plt.annotate("Raiz real", xy=(xbar,0), xytext=(-7.5,5),
             arrowprops=dict(facecolor='black', shrink=0.05),fontsize=15)

plt.annotate("Punto fijo", xy=(xbar,-3.0), xytext=(2.5,-5),
             arrowprops=dict(facecolor='black', shrink=0.05),fontsize=15)




plt.xlabel(r"$x$", fontsize=15)
plt.ylabel(r"$f(x)$",fontsize=15)

plt.legend()
plt.show()
```
